```python
#Média do valor dos imoveis
media_imoveis = df.groupby('Tipo')['Valor'].mean().round(decimals=2)
media_imoveis
```

```python
#Ordenar em ordem crescente
media_imoveis_cresc = media_imoveis.sort_values(ascending=True)
media_imoveis_cresc
```

```python
imoveis_comerciais = ['Indústria','Chácara','Prédio Inteiro','Loja/Salão','Conjunto Comercial/Sala','Galpão/Depósito/Armazém','Box/Garagem','Sítio','Pousada/Chalé','Loteamento/Condominio','Casa Comercial']
```

```python
#query == consulta
df.query('@imoveis_comerciais in Tipo')
```

```python
df_atualizado = df.query('@imoveis_comerciais not in Tipo')
df_atualizado
```

```python
df_imoveis_menor_ao_maior_2 = df_atualizado.groupby('Tipo')[["Valor"]].mean().sort_values("Valor").round(decimals=2)
```

```python
df_imoveis_menor_ao_maior_2.plot(kind='barh', figsize=(14,10), color="red")
```
